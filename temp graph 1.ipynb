First step is to dowload the raw data from: https://www.metoffice.gov.uk/hadobs/hadcrut5/data/current/download.html

```python

# Import Libraries

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Import Data

#df = pd.read_excel ('tempdata3.xlsx')    #r'C:\Users\danho\
df=pd.read_csv('tempdata3.csv')
#df2=pd.read_csv('hadcut_monthly.csv')
```

The Next Step is to add reference numbers and calculate the rolling average at a range of time scales.

```python
# Add Preindustrial and Paris Agreement Targets

df['paris1_5'] = np.full((173,1),1.5)
df['paris2_0'] = np.full((173,1),2.0)
df['preind'] = np.full((173,1),0)

years = df['Year']
temps = df['HadCRUT5 (degC)']
paris1_5 = df['paris1_5']
paris2_0 = df['paris2_0']
preind = df['preind']

# Calculate rolling mean

df['rolling_temps_10']=df['HadCRUT5 (degC)'].rolling(10).mean()
tenyearav = df['rolling_temps_10']
df['rolling_temps_30']=df['HadCRUT5 (degC)'].rolling(20).mean()
thirtyyearav = df['rolling_temps_30']
```

Final Step is to make a figure including all the data.

```python
# plot data
fig, ax = plt.subplots()
ax.plot(years,temps,'bo',markersize=3,label='Annual')      #linewidth=0.8
ax.plot(years,tenyearav,'r-',linewidth=2,label='10 Year Rolling Average')
ax.plot(years,thirtyyearav,'r-',linewidth=4,label='20 Year Rolling Average')
ax.set_title('Global Average Temperature Change relative to PreIndustrial')
ax.set_xlabel('Year')
ax.set_ylabel('Temperature Anomaly')

ax.plot(years,paris1_5,color='black',linestyle='--',label='_nolegend_')
ax.plot(years,paris2_0,color='black',linestyle='--',label='_nolegend_')
ax.plot(years,preind,color='grey',linestyle='--',label='_nolegend_')

ax.legend(edgecolor='white')
plt.legend()

plt.gcf().set_size_inches(10,8)
```
