# CO2 graph

Data from: [Keeling Curve](https://scrippsco2.ucsd.edu/data/atmospheric_co2/primary_mlo_co2_record.html)

```python
# Import Libraries

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Download Data

df = pd.read_csv('monthly_co2.csv')
```

Wrangle the data

```python
# Remove -99.99 values

df = df.drop(df[df['CO2[ppm]'] == -99.99].index)

dates = df['Date']
CO2 = df['CO2[ppm]']
```

Plot the Data

```python

fig, ax = plt.subplots()
ax.plot(dates,CO2,'ro',markersize=1)
ax.set_xlabel('Year')


plt.gcf().set_size_inches(10,8)

plt.savefig('co2.png', bbox_inches='tight')
```
